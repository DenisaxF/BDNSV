```python
from neo4j import GraphDatabase
import pandas as pd

# Configurare conexiune - COMPLETA»ñI CU DATELE VOASTRE
URI = "neo4j+s://72bc61dd.databases.neo4j.io"  # Din Neo4j Aura
AUTH = ("neo4j", "3mfPZTJmplgP0huqxhXz9alrz33KaeN6DmIjKtIaF8E")  # Username »ôi parola voastrƒÉ

driver = GraphDatabase.driver(URI, auth=AUTH)

def run_query(query, parameters=None):
    """Func»õie helper pentru rularea query-urilor"""
    with driver.session() as session:
        result = session.run(query, parameters or {})
        return [record.data() for record in result]

# Test conexiune
try:
    driver.verify_connectivity()
    print("‚úì Conectare reu»ôitƒÉ la Neo4j!")
except Exception as e:
    print(f"‚úò Eroare la conectare: {e}")
```

```python
# Creare constr√¢ngeri »ôi indexuri pentru performan»õƒÉ
constraints = [
    "CREATE CONSTRAINT student_id IF NOT EXISTS FOR (s:Student) REQUIRE s.id IS UNIQUE",
    "CREATE CONSTRAINT profesor_id IF NOT EXISTS FOR (p:Profesor) REQUIRE p.id IS UNIQUE",
    "CREATE CONSTRAINT curs_cod IF NOT EXISTS FOR (c:Curs) REQUIRE c.cod IS UNIQUE",
    "CREATE CONSTRAINT proiect_id IF NOT EXISTS FOR (pr:Proiect) REQUIRE pr.id IS UNIQUE",
    "CREATE INDEX student_nume IF NOT EXISTS FOR (s:Student) ON (s.nume)",
]

for constraint in constraints:
    try:
        run_query(constraint)
    except Exception as e:
        print(f"Constr√¢ngere deja existentƒÉ sau eroare: {e}")

print("‚úì Constr√¢ngeri »ôi indexuri create!")
```

```python
#Exerci»õiul P1: Afi»ôa»õi to»õi profesorii din departamentul de InformaticƒÉ, ordona»õi alfabetic.
query = """
MATCH (p:Profesor)
WHERE p.departament = 'InformaticƒÉ'
RETURN p.nume AS Nume
ORDER BY p.nume ASC
"""
result = run_query(query)
df = pd.DataFrame(result)
print("Profesori Informatica:")
print(df)
```

```python
#Exerci»õiul P2: GƒÉsi»õi toate proiectele cu status '√Æn desfƒÉ»ôurare'.
query = """
MATCH(pr:Proiect)
WHERE pr.status = '√Æn desfƒÉ»ôurare'
RETURN pr.nume AS Proiect
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Proiecte in desfasurare:")
print(df)
```

```python
#Exerci»õiul P3: Afi»ôa»õi studen»õii cu media peste 9.0.

query = """
MATCH(s:Student)
WHERE s.medie > 9.0
RETURN s.nume AS Nume, s.medie AS Medie
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Studenti cu media peste 9.0:")
print(df)
```

```python
#Exerci»õiul P4: Lista»õi toate cursurile care au mai mult de 5 credite.
query = """
MATCH(c:Curs)
WHERE c.credite > 5
RETURN c.nume AS Curs, c.credite AS Nr_credite
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Cursuri cu mai mult de 5 credite:")
print(df)
```

```python
#Exerci»õiul P5: Pentru fiecare proiect, afi»ôa»õi studen»õii care lucreazƒÉ la el »ôi rolul lor.

query = """
MATCH(s:Student)-[w:WORKS_ON]->(pr:Proiect)
RETURN pr.nume AS Proiect, s.nume as Student , w.rol AS Rol
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Studenti si proiexte:")
print(df)
```

```python
#Exerci»õiul P6: GƒÉsi»õi studen»õii care au note peste 9.0 la toate cursurile la care sunt √Ænscri»ôi.
query = """
MATCH(s:Student)-[e:ENROLLED_IN]->(c:Curs)
WHERE s.medie > 9.0
RETURN DISTINCT s.nume AS Nume
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Studenti cu media peste 9.0:")
print(df)
```

```python
#Exerci»õiul P9: Afi»ôa»õi pentru fiecare profesor lista de studen»õi pe care √Æi mentoreazƒÉ.

query = """
MATCH(s:Student)-[m:MENTORED_BY]->(p:Profesor)
RETURN p.nume AS Profesor, collect(s.nume) AS Studen»õi
"""

result = run_query(query)
df = pd.DataFrame(result)
print("Profesori si studenti:")
print(df)
```
